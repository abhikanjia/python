### 02) WAP to to Plot List random of X, Y Coordinates in Matplotlib.

```python
import random 
xlist=[]
ylist=[]
for i in range(1,11):
    a=random.randint(1,10)
    b=random.randint(1,10)
    xlist.append(a)
    ylist.append(b)
print(xlist)
print(ylist)
plt.plot(xlist,ylist)
plt.show()
```

### 03) WAP to demonstrate the use of Bar chart.

```python
x = [8,9,6,3,5]
y = [5,8,9,6,3]
l =['1st','2nd','3rd','4th','5th']
c =['b','g','r','c','m']
plt.title('Marks')
plt.bar(x,y,color=c,label=l)
for i in range(len(y)):
    plt.text(x[i],y[i],y[i])
plt.show()
```

### 04) WAP to demonstrate the use of Histogram.

```python
import random
cpi=[]
for i in range(100): 
    a=random.randint(0,10)
    cpi.append(a)
plt.hist(cpi,bins=10, histtype='stepfilled',align='mid',label='CPI Hist')
plt.legend()
plt.show()
```

### 01) WAP to display the value of each bar in a bar chart using Matplotlib.

```python
x = ["python","c","c++","java"]
y = [85,70,60,82]
c = ['r','g','y','m']
plt.bar(x,y,color=c,edgecolor="b")
def addLabels(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i])
plt.title("popularity of languages")
plt.xlabel("languages")
plt.ylabel("ratings")
addLabels(x,y)
plt.show()
```

### 02) WAP create a Scatter Plot with several colors in Matplotlib?

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]
c = ["r","b","g","r","b"]
width = [1,2,3,4,5]
plt.scatter(x,y,color=c,marker="*",linewidths=width)
plt.xlabel("x")
plt.ylabel("y")
plt.title("star")
plt.show()
```
